```python
import os
import random
import matplotlib.pyplot as plt
from collections import Counter
from torchvision.datasets import ImageFolder
from torchvision import transforms
```

```python
train_dir = "dataset\\rafdb\\train"
test_dir = "dataset\\rafdb\\test"
```

```python
train_dataset = ImageFolder(train_dir)
test_dataset = ImageFolder(test_dir)

print("Class order:", train_dataset.classes)
print("Class to idx:", train_dataset.class_to_idx)
```

```python
train_labels = [label for _, label in train_dataset.samples]
test_labels = [label for _, label in test_dataset.samples]

train_counts = Counter(train_labels)
test_counts = Counter(test_labels)

print("Train distribution:", train_counts)
print("Test distribution:", test_counts)
```

```python
plt.figure(figsize=(10,5))

plt.bar(train_counts.keys(), train_counts.values())
plt.title("Train Class Distribution")
plt.xlabel("Class Index")
plt.ylabel("Count")

plt.show()
```

```python
plt.figure(figsize=(12,8))

class_indices = train_dataset.classes

for i in range(len(class_indices)):
    idx = train_labels.index(i)
    img_path, label = train_dataset.samples[idx]
    
    img = plt.imread(img_path)

    plt.subplot(2,4,i+1)
    plt.imshow(img)
    plt.title(f"Class {class_indices[i]}")
    plt.axis("off")

plt.tight_layout()
plt.show()
```

```python
print("=== INSIGHT ===")

total = sum(train_counts.values())

for k, v in sorted(train_counts.items()):
    print(f"Class {k}: {v} ({v/total:.2%})")
```
